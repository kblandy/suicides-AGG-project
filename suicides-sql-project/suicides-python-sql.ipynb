```python
import os
import pandas as pd
import sqlite3

#Creates a database (in this case storing into local memory)
db = sqlite3.connect('country-suicides.db')

#read file
df1 = pd.read_csv('master.csv') 

# convert the pandas dataframe to a sqlite table
df1.to_sql("suicides", db, if_exists="replace")

#returns query in table format
def run_query(query):
    return pd.read_sql_query(query, db)

# Change column names
query = """
    SELECT country AS Country, 
           year AS Year,
           sex AS Sex,
           suicides_no AS [Number of Suicides],
           population AS Population,
           generation AS Generation
    FROM suicides
    LIMIT 100
"""

run_query(query)

```

```python
#shows number of suicides, average suicides, and generation for each generation
query = """
    
        SELECT SUM(suicides_no) AS [Suicides from year 1987-2014],
               ROUND(AVG(suicides_no), 2) AS [Average Suicides per year],
               generation AS Generation
        FROM suicides
        WHERE generation = "Generation X"
        
        UNION ALL
        
        SELECT SUM(suicides_no) AS [Suicides from year 1987-2014],
               ROUND(AVG(suicides_no), 2) AS [Average Suicides per year],
               generation AS Generation
        FROM suicides
        WHERE generation = "Silent"
        
        UNION ALL
        
        SELECT SUM(suicides_no) AS [Suicides from year 1987-2014],
               ROUND(AVG(suicides_no), 2) AS [Average Suicides per year],
               generation AS Generation
        FROM suicides
        WHERE generation = "G.I. Generation"
        
        UNION ALL
        
        SELECT SUM(suicides_no) AS [Suicides from year 1987-2014],
               ROUND(AVG(suicides_no), 2) AS [Average Suicides per year],
               generation AS Generation
        FROM suicides
        WHERE generation = "Boomers"
        
        UNION ALL
        
        SELECT SUM(suicides_no) AS [Suicides from year 1987-2014],
               ROUND(AVG(suicides_no), 2) AS [Average Suicides per year],
               generation AS Generation
        FROM suicides
        WHERE generation = "Millenials"
        
"""

run_query(query)
```
